````python
# Импорт модуля или модулей целиком
import math
cos_value = math.cos(math.pi/25)
sin_value = math.sin(math.pi/25)
result = cos_value**2 * sin_value**2
print(round(result, 1))

# Импорт модуля с присвоением псевдонима модулю
import math as m
cos_value = m.cos(m.pi/25)
sin_value = m.sin(m.pi/25)
result = cos_value**2 * sin_value**2
print(round(result, 1))

# Импорт конкретного объекта из модуля
from math import cos, sin, pi
cos_value = cos(pi/25)
sin_value = sin(pi/25)
result = cos_value**2 * sin_value**2
print(round(result, 1))

# Импорт конкретного объекта из модуля с присвоением псевдонима объекту
from math import cos as c, sin as s, pi as p
cos_value = c(p/25)
sin_value = s(p/25)
result = cos_value**2 * sin_value**2
print(round(result, 1))

# Импорт всех объектов из модуля
from math import *
cos_value = cos(pi/25)
sin_value = sin(pi/25)
result = cos_value**2 * sin_value**2
print(round(result, 1))
````
