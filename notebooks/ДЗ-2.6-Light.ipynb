Посчитать значение $\cos^2(\pi/7)$ и $\sin^2(\pi/7)$, используя указанные функции из библиотеки ```math```, применив для этого **каждую** из конструкций импорта:
``` python
# 1. Импорт модуля или модулей целиком
import имя_модуля
import имя_модуля1, имя_модуля2

# 2. Импорт модуля с присвоением псевдонима модулю

# 3. Импорт конкретного объекта из модуля

from имя_модуля import имя_объекта1, имя_объекта2

# 4. Импорт конкретного объекта из модуля с присвоением псевдонима объекту
```

Найдите сумму $\cos^2(\pi/7)$ и $\sin^2(\pi/7)$.

**Формат ответа:** Число с плавающей запятой, округленное до десятых (например,```0.1```)

**Решение:**

````python
# Импорт модуля или модулей целиком
import math
cos_value = math.cos(math.pi/7)
sin_value = math.sin(math.pi/7)
result = round(cos_value**2 + sin_value**2, 1)
print(result)

# Импорт модуля с присвоением псевдонима модулю
import math as m
cos_value = m.cos(m.pi/7)
sin_value = m.sin(m.pi/7)
result = round(cos_value**2 + sin_value**2, 1)
print(result)

# Импорт конкретного объекта из модуля
from math import cos, sin, pi
cos_value = cos(pi/7)
sin_value = sin(pi/7)
result = round(cos_value**2 + sin_value**2, 1)
print(result)

# Импорт конкретного объекта из модуля с присвоением псевдонима объекту
from math import cos as c, sin as s, pi as p
cos_value = c(p/7)
sin_value = s(p/7)
result = round(cos_value**2 + sin_value**2, 1)
print(result)

# Импорт всех объектов из модуля
from math import *
cos_value = cos(pi/7)
sin_value = sin(pi/7)
result = round(cos_value**2 + sin_value**2, 1)
print(result)


````
